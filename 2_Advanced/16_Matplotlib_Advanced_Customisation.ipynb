```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import datetime as dt
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Code from exercise 12

df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()

df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA_explode = df_DA.explode('job_skills')

df_DA_pivot = df_DA_explode.pivot_table(index = 'job_posted_month_no', columns = 'job_skills', aggfunc = 'size', fill_value = 0)

df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot.sort_values(axis = 1, by = 'Total', inplace= True, ascending= False)

df_DA_pivot.drop('Total', inplace=True)

df_DA_pivot_short = df_DA_pivot.iloc[:, 0:5]

df_DA_pivot_short.plot(
    kind = 'line',
    linewidth = 4,
    linestyle = ':', # options: '-', '--', '-.', ':'
    colormap = 'viridis', # options: check cheatsheet
    marker = 'o',
    markersize = 5,
    figsize = (10, 5)
    
) # just check matplotlib documentation

plt.title('Top 5 Skills for Data Analysts per month')
plt.xlabel('')
plt.ylabel('Count')
plt.show()

df_DA_pivot
```

```python
from adjustText import adjust_text
df_exploded = df.explode('job_skills')
skill_stats = df_exploded.groupby('job_skills').agg(
    median_salary=('salary_year_avg', 'median'),
    skill_count=('job_skills', 'count')
)
count = 15

skill_stats = skill_stats.sort_values(by = 'skill_count', ascending = False).head(count)

skill_stats.plot(kind = 'scatter', x = 'skill_count', y = 'median_salary')

texts = []
for i, txt in enumerate(skill_stats.index):
    texts.append(plt.text(skill_stats['skill_count'].iloc[i], skill_stats['median_salary'].iloc[i], txt))

adjust_text(texts, arrowprops = dict(arrowstyle = '->', color = 'gray', lw = 0.5)) # Meant to space out scatter labels


ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K'))
plt.xlabel('Count of Job Postings')
plt.ylabel('Median of Yearly Salary ($USD)')
plt.title(f'Salary vs. Count of Job Postings for Top {count} Skills')
plt.tight_layout()
plt.show()
```
